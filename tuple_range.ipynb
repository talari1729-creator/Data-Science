```python
## Tuples are immutable(unchangeable),ordered sequence of items.
## they are defined uding parenthesis(),with separated by commas.
## like lists they can contain a different data types.
```

```python
## LIke list tuple also supports the creating,indexing,slicing,len(),membeeship testing(in,not in).
my_info=("name","age","place","city")
print(my_info)
```

```python
my_info
print(len(my_info))
print(my_info[0])
print(my_info[:])
print(my_info[0:3])
print(my_info[:-3])
```

```python
## creating a single element tuple( note the comma)
single_tool=("raajan",)
print(single_tool)## output includes comma also.
```

```python
## tuple packing ( creating a tuple without parenthesis)
my_info="name","age","place"
print(my_info)
```

```python
## tuple.count(item)
## returns the no.of times the specified item appeared in the tuple.
my_family=("amma","nanna","chinnodu","peddodu","amma","nanna","1","2","3","1","3")
print(my_family.count("amma"))
print(my_family.count("3"))
print(my_family.count("raajan"))
```

```python
## tuple.index(item,start,end)
## returns the index of the first occurrence of the specified item.
my_family
nanna_index=my_family.index("nanna")
print(nanna_index)
index_3=my_family.index("3")
print(index_3)
nanna_2=my_family.index("nanna",2)
print(nanna_2)
print(len(my_family))
```

```python
## Range 
## represents the immutable sequence of numbers,typically used for looping a specific no.of times in for loops.
## Range supports the len(),membership testing(in,not in),indexing and slicing.

```

```python
## creating and iteration
## creating range objects snd using them in for loops.
simple_=range(0,6)
print(simple_)
for i in range(0,6):
    print(i,end=" ")
```

```python
for batch_id in range(101,107):
    print(batch_id,end=" ")
```

```python
for i in range(0,6):
    print(i,end=" ")
```

```python
for batch in range(101,107):
    print(batch,end=" ")
```

```python
for even_ in range(0,12,2):
    print(even_,end=" ")
```

```python
for i in range(5,0,-1):
    print(i,end=" ")
```

```python
course_list=["DS","DE","ADO"]
for i in range(len(course_list)):
    print(i,end=" ")
```

```python
for i in range(10,0,-1):
    print(i,end=" ")

 
```

```python
## getting no outputs


for i in range(5,5):
    print(i,end=" ")

for i in range(10,0,1):
    print(i)

for i in range(0,10,-1):
    print(i)

```

```python
## converting the range to list
for i in range(11,18):
    print(i,end=" ")

batch=list(range(11,18))
print(batch)
print(len(batch))
```

```python
## converting range to the tuple
batch=tuple(range(43,49))
print(batch)
```

```python
## length and membershiping
my_list=range(10,25)
print(my_list)
print(len(my_list))

is_20=20 in my_list
print(is_20)

not_30=30 in my_list
print(not_30)

not_30=30 not in my_list
print(not_30)
```
