```python
## list,pop([index])
## removes and return the item at the given index.
## if index is omitted,it removes and returns the last item.
my_sports=["basket ball","cricket","hand ball","kabaddi","throw ball"]

pop_index=my_sports.pop()
print(my_sports)

my_sports.pop(0)
print(my_sports)

my_sports.pop(-1)
print(my_sports)
```

```python
## sort the item of the  list in place.
## key: A function to be called on each element before making comparisons (e.g., key=str.lower for case-insensitive sort).
## reverse=True: Sorts in descending order. Items must be comparable (e.g., all numbers or all strings)

course_list = ["Data Engineering", "Azure DevOps", "Data Science", "Python Basics"]
course_list.sort()
print(course_list)
```

```python
## in numaric cases
student_counts = [350, 420, 380, 500, 450]
student_counts.sort()
print(student_counts)

student_counts.sort(reverse=True)
print(student_counts)
```

```python
## reverse=true
course_list = ["Data Engineering", "Azure DevOps", "Data Science", "Python Basics"]
course_list.sort(reverse=True)
print(course_list)
```

```python
course_list.sort(key=str.lower)
print(course_list)
```

```python
## list of tuple based on the second element
my_score=[("maths",200),("statistics",150),("computer",300)]
my_score.sort(key=lambda item:item[1])
print(my_score)

## reverse
my_score=[("maths",200),("statistics",150),("computer",300)]
my_score.sort(key=lambda item:item[1],reverse=True)
print(my_score)
```

```python
## sorting based on length of the strings.
courses = ["SQL", "Python", "Azure DevOps", "Gits"]
courses.sort(key=len)
print(courses)

courses.sort(key=len,reverse=True)
print(courses)
```

```python
## list.clear()
## removes all the items in a list.
course_list = ["Data Science", "Data Engineering", "Azure DevOps"]
tasks_to_clear = ["Task A", "Task B"]

course_list.clear()
print(course_list)

tasks_to_clear.clear()
print(tasks_to_clear)

numbers=[1,2,3,4,5]
numbers.clear()
print(numbers)

```

```python
## membership testing(in,not in)
## checks if an item exists within the list.Returns True or False.
course_list = ["Data Science", "Data Engineering", "Azure DevOps"]

is_exist="Data Science" in course_list
print(is_exist)

is_not_exist="data analyst" in course_list
print(is_not_exist)
```
