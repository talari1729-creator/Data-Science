```python

# Question 6: Check if 97 is a prime number
num=97
if num % 2 !=0:
    print("True")
else:
    print("False")
```

```python
num=97
is_prime=True

for i in range(2,num):
    if num % i == 0:
        is_prime=False
        break
print("prime" if is_prime else "not prime")
```
