```python
## Python Built-in Numeric Data Type
## (1)=int(integer)
## int (Integer): Represents whole numbers (positive, negative, or zero) without a fractional part.
## Explanation: Integers represent whole numbers. They are fundamental for counting, indexing, and representing discrete quantities like the number of students, courses, or batches.
```

```python
## addition(+)
## adding 2 numbers
num_students=120
num_teachers=15
total=num_students+num_teachers
print(total)
```

```python
## Subtraction (-)
## subtracting the first from the second
num_teachers=15
num_students=120
diff_=num_teachers-num_students
print(diff_)
diff_1=num_students-num_teachers
print(f"different between:{diff_1}")
```

```python
## Multiplication (*)
## multiples the two numbers
num_teachers=15
num_students=120
multi_=num_teachers*num_students
print(multi_)
num_teachers=-15
num_students=120
neg_=num_teachers*num_students
print(neg_)
```

```python
## Floor Division (//)
##  Divides the first number by the second and rounds the result down to the nearest whole number (integer)
## quotient is the answer
total_students = 157
students_per_batch = 50
full_batches=total_students//students_per_batch
print(full_batches)

## Divide by large number  
producers=100
consumers=1000
result=producers//consumers
print(result)

teachers=15
students=170
get_mode=teachers//teachers
print(get_mode)
```

```python
## Modulo (%)
## it return the remainder of the division of the first number by the second.

students=159
teachers=15
results=students % teachers
print(results)

consumers=174
producers=22
remain_consumers=consumers % producers
print(remain_consumers)

student_id=1729
is_odd=student_id % 2
print(f" yes this is my id with odd:{is_odd}")  ## 1 means yes
```

```python
## Exponentiation(**)
## Raises thr first num to the power of second num

growth=2
periods=3
projected_students=(growth ** periods)*100
print(projected_students)

length=4
breadth=2
area=length ** breadth
print(area)

length=4
breadth=4
area=length ** breadth
print(area)

## anything power with zero,output is 1

base=-5
power=2
result=base ** power
print(result)
base=-5
power=3
result=base ** power
print(result)

```

```python
## Absolute value
## abs()...returns the absolute (non negative) value of a num

student_id=-1729
result=abs(student_id)
print(result)

length=30
breadth=15
diff_=abs(breadth - length)
print(diff_)
```

```python
## pow(base,exp[mod])
## it returns the exponentation (x ** y)

result=pow(2,3)
print(result)

side_length=5
power=2
area=pow(side_length,power)
print(area)

## if we add modulus
side_length=5
power=3
area=pow(side_length,power,2)
print(area)
```

```python
## divmod(a,b)
## it returns the value along with the quotient(div) and remainder(mod)
total_students=146
teachers=15
results=divmod(total_students,teachers)
print(results)

assignments=47
instructors=4
assign_per=divmod(assignments,instructors)
print(assign_per)
```

```python
## int()
## while also the constructor,can be used as a function to convert the values from the other types ( like float or compatible string and boolean) to an interger.
## it avoids the decimals

avg_rating=4.5
print(int(avg_rating))

student_str="1729"
print(int(student_str))

neg_float_str="-1729"
print(int(neg_float_str))
raman=-0.1729
print(int(raman))

active_true=True
print(int(active_true))
active_false=False
print(int(active_false))
```

```python
## float pointing numbers
## float,it represents the values with a decimal points,not be a whole number.
## float support the same arithmetic operations(+,-,*./),but results in float.

## floor division(//) and modulus(%),this operations also work with float.(output should be absolute float value)

### if you want float values as a output you must provide input values as float.

avg_rate=4.6
total_points=1.0
result=avg_rate//total_points
print(result)
```

```python
## round(num , ndigits)
##  Rounds a number to a specified number of decimal places (ndigits).

avg_rating=4.45678
print(round(avg_rating))
print(round(avg_rating,0))
print(round(avg_rating,1))
print(round(avg_rating,2))
print("round")
avg_val=49.94567
print(round(avg_val))
print(round(avg_val,0))
print(round(avg_val,1))
```

```python
## rounding to negative value
neg_val=-4.57
print(round(neg_val))
print(round(neg_val,0))
print(round(neg_val,1))
print(round(neg_val,2))
```

```python
## float.is_integer()
## in this function, if there is a value after the decimal point,it returns the False.
## if there is no value after the decimal,except (0),it returns the True.
avg_rating=4.8
print(avg_rating.is_integer())

avg_val=5.00
print(avg_val.is_integer())

side=10.0
length=3.0
div=side/length
print(div.is_integer())
```

```python
## float.as_integer_ratio()
## returns the pair of integer values with in a tuple to the decimal value.

avg_rate=4.5
print(avg_rate.as_integer_ratio())

ratio=0.5
18
print(ratio.as_integer_ratio())

## for nagative
ratio_val=-4.67
print(ratio_val.as_integer_ratio())
```

```python
## float()
## the type constructor,can also be used as a function to convert the float from the other types(int or compatable strings).

students=1500
teachers=200
total=float(students+teachers)
print(total)

zero_=float(0)
print(zero_)

sci_str="1.54e3"
print(float(sci_str))

true="True"
print(float(true))  ##could not convert string to float: 'True
```

```python
## Text type (str)-------strings are immutable.
## strings are used to represents the textual data with in single,double and triple quotes.
## triple quotes in multiple-line strings or docstrings.
## examples
institute='sv university'
branch="science"
city='''tirupati'''
```

```python
## Concatination(+)-joins 2 strings together,creating a new string.


institute="sv.university"
course="m.sc.statistics"
result=institute+" : "+course
print(result)

place="tirumala"
god="balaji"
result="welcome to "+place+" to visit "+god
print(result)
```

```python
## repetition(*)-repeats the string a specified no.of times,creating a new string.
repeat="om namo venkatesaya"
result=repeat*1000
print(result)
```

```python
## len()
## it is a function to find the length(no.of characters) in a string.
word="om namo venkatesaya"
result=len(word*1000)
print(result)
```

```python
## indexing( [] )
## accessing a single char at a specifc position(index)
## python uses a zero based indexing(first char)
## negative indices count fron the end(-1 ,is the last char)

university="sv.university"
place="tirupathi"
result=university+" "+place
print(result)
length=len(result)
print(length)
print(result[0])
print(result[18])
print(result[-8])
```

```python
## Slicing(  [:]  )
## extracting a portion in a string using [start:stop:step]
## start-the index where the slice begins....(inclusing)
## stop-the index where the slice end.....(excluding)
## step-the increament between indices
```

```python
insti="edukron"
place="bangaluru"
result=insti+" "+place
print(result)
length=len(result)
print(length)
print(result[12])
print(result[:])
print(result[8:14])
print(result[:-8])
print(result[-11:])
print(result[0:15:2])
```

```python
## str.upper
## returns a new string with all chars converted in upper case
university="sri venkateswara university"
result_upper=str.upper(university)*5
print(result_upper)
```

```python
## str.lower()
## return the new string with all chars are cnverted in to lower case.
university="SRI VENKATESWARA UNIVERSITY"
result_lower=str.lower(university)
print(result_lower)
```
