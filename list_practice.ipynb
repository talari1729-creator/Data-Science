```python
## Creating and indexing/slicing
course_tools=["python","SQL","powerbi","pandas","numpy"]
print(course_tools)
print(len(course_tools))
```

```python
## indexing
print(course_tools[0])
print(course_tools[-1])
print(course_tools[:])
```

```python
## slicing
course_tools
print(course_tools[-4])
print(course_tools[-3:3])
print(course_tools[0:-1])
print(course_tools[1:5:2])
```
