```python
class bankaccount:

    def __init__(self,account_num,account_holder,balance):
        self.account_num=account_num
        self.account_holder=account_holder
        self.balance=balance
    
    def display(self):
        print(f"account_num:{self.account_num},name:{self.account_holder,balance:{self.balance}}")

    def deposite(self,amount):
        global balance
        balance+=amount
    print(f"deposited:{amount},new balance:{self.balance}")

    def withdraw(self,amount):
        global balance
        if amount > balance:
            print(f"withdraw: ${amount}.new balance:{self.balance}")
        else:
            balance-=amount
    print(f"updated balance:",balance)

    def check_balance(self):
        print(f"current balance:{self.balance}")
```

```python
class library:
    def __init__(self,books):
        self.books=books
        
    def add_book(self,book_name,title,author):
        book={"book_name":"book_name",
              "title":"title",
              "author":"author"
              }
        self.books.append(book)
    
    def remove_book(self,book_name):
        for b in self.books:
            if b["book_name"]==book_name:
                self.books.remove(b)

    def display_books(self):
        for book in self.books:
            print(book)

    def find_book(self,book_name):
        for book in self.books:
            if book["book_name"]==book_name:
                return book
            return None
```
