```python

# Question 10: Create a list of prime numbers between 1 and 50
primes = []
for num in range(2, 51):   # start from 2 since 1 is not prime
    is_prime = True
    for i in range(2, int(num**0.5) + 1):  # check divisors up to sqrt(num)
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print(primes)
```
