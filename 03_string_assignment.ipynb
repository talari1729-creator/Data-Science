```python
# Question 11: Find the most frequent character in "mississippi"
word="mississippi"
freq={}
for char in word:
    freq[char]=freq.get(char,0)+1

most_freq=max(freq,key=freq.get)
print(most_freq,freq[(most_freq)])
```
