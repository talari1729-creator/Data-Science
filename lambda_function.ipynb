## lambda ia a anonymous function and it is one liner function

```python
## regural function
def add(a,b):
    return a+b
```

```python
print(add(2,5))
```

```python
## lambda function
addition=lambda a,b:a+b
```

```python
print(addition(2,5))
```

```python
square=lambda x:x**2
print(square(3))
print(square(5))
print(square(8))
```

```python
## lambda with multiple arguments
area=lambda length,breadth:length*breadth
print(area(5,4))
```

```python
## lambda with condition
## check if even or odd
is_even=lambda x : "even" if x % 2==0 else "odd"
print(is_even(67))
print(is_even(98))
```

```python
is_even(78)
```

```python
## check negative or positive
is_negative=lambda x:"positive" if x>0 else "negative"
print(is_negative(-4))
print(is_negative(8))
```

```python
## check pass or fail
is_pass=lambda x:"pass" if x>=35 else "fail"
print(is_pass(35))
print(is_pass(87))
print(is_pass(-76))
```

```python
## check adult or minor
is_minor=lambda x:"minor" if x<=18 else "adult"
print(is_minor(66))
print(is_minor(18))
```

##  map function with lambda using list of numbers
## map function allpy to all elements in a list

```python
## regular function without lambda
def square(x):
    return x*x
nums=[1,2,3,4,5]
result=list(map(square,nums))
print(result)
```

```python
## using lambda
nums=[2,3,4,5]
result2=list(map(lambda x:x*x,nums))
print(result2)
```

```python
nums=[2,3,4,5]
result2=list(map(lambda x:x*5,nums))
print(result2)
```

```python
## check true or false while using map
nums=[2,3,4,5]
result2=list(map(lambda x:x%2==0,nums))
print(result2)
```

## filter function apply for only specific conditions in a list

```python
nums=[2,3,4,5]
result2=list(filter(lambda x:x%2==0,nums))
print(result2)
```

## using sorted


```python
students=[("raju",70),("buchi",80),("nanna",90)]
print(sorted(students))
```

```python
result=sorted(students,key=lambda x:x[0])
print(result)
```

```python
employees=[
    ("raju",5000,5),
    ("buchi",4000,4),
    ("nanna",3000,3)
]
```

```python
result=sorted(employees,key=lambda x:(x[1],x[2]))
print(result)
```

```python
result2=list(filter(lambda x:x[1]>=6000 and x[2]<=4,employees))
print(result2)
```
